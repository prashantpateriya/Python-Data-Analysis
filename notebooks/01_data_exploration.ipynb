# Data Exploration Notebook

This notebook demonstrates basic data exploration techniques using pandas and other Python libraries.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scripts.data_loader import load_csv, get_data_info, clean_data

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("Libraries imported successfully!")
```

## Load Sample Data

For this demo, we'll create a sample dataset since we don't have real data files yet.

```python
# Create sample dataset
np.random.seed(42)
n_samples = 1000

data = {
    'id': range(1, n_samples + 1),
    'age': np.random.normal(35, 12, n_samples).astype(int),
    'income': np.random.lognormal(10, 0.5, n_samples),
    'score': np.random.uniform(0, 100, n_samples),
    'category': np.random.choice(['A', 'B', 'C', 'D'], n_samples),
    'active': np.random.choice([True, False], n_samples, p=[0.7, 0.3])
}

df = pd.DataFrame(data)
print(f"Dataset created with {len(df)} rows and {len(df.columns)} columns")
df.head()
```

## Basic Data Information

```python
# Get basic information about the dataset
info = get_data_info(df)

print("=== Dataset Information ===")
print(f"Shape: {info['shape']}")
print(f"Memory usage: {info['memory_usage'] / 1024:.2f} KB")
print(f"\nColumns: {info['columns']}")
print(f"\nNumeric columns: {info['numeric_columns']}")
print(f"Categorical columns: {info['categorical_columns']}")
```

```python
# Check for missing values
missing_values = info['missing_values']
print("Missing values:")
for col, count in missing_values.items():
    if count > 0:
        print(f"  {col}: {count}")
    else:
        print(f"  {col}: No missing values")
```

## Statistical Summary

```python
# Statistical summary of numeric columns
print("Statistical Summary:")
df[info['numeric_columns']].describe().round(2)
```

## Data Visualization

```python
# Create subplots for distribution plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribution of Numeric Variables', fontsize=16)

# Age distribution
axes[0, 0].hist(df['age'], bins=20, alpha=0.7, color='skyblue')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')

# Income distribution
axes[0, 1].hist(df['income'], bins=20, alpha=0.7, color='lightgreen')
axes[0, 1].set_title('Income Distribution')
axes[0, 1].set_xlabel('Income')
axes[0, 1].set_ylabel('Frequency')

# Score distribution
axes[1, 0].hist(df['score'], bins=20, alpha=0.7, color='salmon')
axes[1, 0].set_title('Score Distribution')
axes[1, 0].set_xlabel('Score')
axes[1, 0].set_ylabel('Frequency')

# Category counts
df['category'].value_counts().plot(kind='bar', ax=axes[1, 1], color='purple', alpha=0.7)
axes[1, 1].set_title('Category Distribution')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Count')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Correlation Analysis

```python
# Correlation matrix for numeric variables
numeric_df = df[info['numeric_columns']]
correlation_matrix = numeric_df.corr()

print("Correlation Matrix:")
print(correlation_matrix.round(3))

# Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.3f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## Data Cleaning Example

```python
# Introduce some missing values for demonstration
df_with_missing = df.copy()
df_with_missing.loc[np.random.choice(df.index, 50), 'age'] = np.nan
df_with_missing.loc[np.random.choice(df.index, 30), 'income'] = np.nan

print(f"Missing values before cleaning:")
print(df_with_missing.isnull().sum())

# Clean the data
df_clean = clean_data(df_with_missing, drop_na=False, fill_na={'age': df['age'].median(), 'income': df['income'].median()})

print(f"\nMissing values after cleaning:")
print(df_clean.isnull().sum())
```
